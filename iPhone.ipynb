# 3. Nhắc lại thuật toán SVM

image.png

Nhìn vào hình trên ta có những thứ sau: hai nhãn riêng biệt (các điểm mãu xanh và các điểm màu đỏ), vậy svm hoạt động như nào ?

+ svm sẽ cố tạo ra một đường (đối với không gian hai chiều) hoặc mặt phẳng phẳng (không gian ba chiều) hoặc siêu phẳng (không gian lớn hơn 3 chiều) để có thể phân cách được các nhãn, trong trường hợp này ta đang là không gian hai chiều nên svm sẽ cố tạo ra một đường thẳng. đường thẳng này sẽ có điều
+ đường thẳng phân chia sẽ được mô hình học sau cho khoảng cách từ nó đến các điểm của các nhãn là lớn nhất.
+ các điểm gần đường thẳng phân chia nhất gọi là các vec tơ hỗ trợ, như vậy để có được đường thẳng phân chia mô hình cũng phải tìm đồng thời các điểm gần đường thẳng gần nhất, tức là thực hiện đồng thời hai việc là tối đa khoảng cách giữa hai lớp và tối thiểu khoảng cách của các điểm gần nhất của hai nhãn đến đường thẳng phân chia.
+ Như ta thấy chỉ các điểm gần nhất đường phân chia (vec tơ hỗ trợ) mới đóng góp vào kết quả của thuật toán do đó ngay cả khi ta loại bỏ các điểm khác (điểm ở xa đường thẳng) cũng sẽ không có sự thay đổi về kết quả.

**SVM có gì đặc biệt ?**


Giả sử ta đang xây dựng một thuật toán để phân loại táo và cao. Điều mà hầu hết các thuật toán  học máy sẽ làm là sẽ xem xét các loại táo trông phổ biến nhất và loại cam phổ biến hất để tự học và đào tạo, dựa vào đó mô hình sẽ xác định các mẫu mới là táo hay cam.

Đối với SVM thì lại khác, thay vì xem xét hầu hết các loại táo và cam thông thường, svm sẽ xem xét những quả táo giống với cam và những quả cam giống với táo.

Nhìn vào hình dưới, svm sẽ chọn quả táo bên trái trông rất giống quả cam và sẽ chọn quả cam màu xanh lá cây ở bên phải trông rất giống táo xanh. Vì vậy, 2 quả này (điểm dữ liệu này) sẽ đại diện cho các vec tơ hỗ trợ và 2 quả này là hai quả gần nhất với danh giới phân biệt khác nhau giữa hai loại quả. Vì vậy svm là một loại thuật toán rất khác vì nó chọn trường hợp đặc biệt để gần ranh giới phân loại nhất và sẽ sử dụng trường hợp này để xây dựng phân tích. Đó là lý do tại sao trong một số trường hợp nhất định, svm hoạt động tốt hơn các thuật toán phân loại khác.

# I. Import thư viện và tải dữ liệu

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import seaborn as sns


dataset = pd.read_csv("./iphone.csv")
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, 3].values
```

# II. Data preparation

```python
labelEncoder_gender =  LabelEncoder()
X[:,0] = labelEncoder_gender.fit_transform(X[:,0])
X = np.vstack(X[:, :]).astype(np.float)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Step 4: Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# III. Modeling & Evaluation

```python
# Step 5 - modeling

# classifier = LogisticRegression(random_state=0, solver="liblinear")
# classifier = KNeighborsClassifier(n_neighbors=5)
# classifier = SVC(kernel = "linear", random_state=0)
# classifier = SVC(kernel ='rbf')
# classifier = GaussianNB()
# classifier = DecisionTreeClassifier(criterion = "entropy", random_state=0)
classifier = RandomForestClassifier(n_estimators=10, criterion="entropy", random_state=0)

classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
cm = metrics.confusion_matrix(y_test, y_pred)
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy score:",accuracy)
precision = metrics.precision_score(y_test, y_pred)
print("Precision score:",precision)
recall = metrics.recall_score(y_test, y_pred)
print("Recall score:",recall)
```
